```python
geral = player_stats[['player', 'position', 'team', 'age', 'assists' , 'goals', 'cards_yellow', 'cards_red',]].copy()
geral['minutes_per_game'] = playtime['minutes_per_game']
geral[['touches', 'dribbles', 'dribbles_completed', 'passes_received']] = possesion[['touches', 'dribbles', 'dribbles_completed', 'passes_received']]
geral['passes_completed'] = passing['passes_completed']
geral[['fouls', 'fouled']] = misc[['fouls', 'fouled']]
```

```python
'FW' # crosses

fw = geral.query('position == "FW"').copy()
fw[['xg', 'shots','shots_on_target', 'shots_on_target_pct', 'goals_per_shot', 'goals_per_shot_on_target']] = shooting[['xg', 'shots','shots_on_target', 'shots_on_target_pct', 'goals_per_shot', 'goals_per_shot_on_target']]
fw['gca'] = gca['gca']
fw['crosses'] = misc['crosses']
fw
```

```python
'MF' # interceptions, xg_assist (?)

mf = geral.query('position == "MF"').copy()

mf[['passes_completed_short', 'passes_short', 'passes_pct_short', 'passes_completed_medium', 'passes_medium', 'passes_pct_medium', 'passes_completed_long', 'passes_long', 'passes_pct_long', 'progressive_passes' ]] = passing[['passes_completed_short', 'passes_short', 'passes_pct_short', 'passes_completed_medium', 'passes_medium', 'passes_pct_medium', 'passes_completed_long', 'passes_long', 'passes_pct_long', 'progressive_passes' ]]

mf['crosses'] = misc['crosses']
mf['gca'] = gca['gca']
mf[['tackles', 'tackles_won', 'blocked_passes', 'interceptions']] = defense[['tackles', 'tackles_won', 'blocked_passes', 'interceptions']]
mf['xg_assist'] = player_stats['xg_assist']
mf
```

```python
'DF' #  , ball_recoveries, pens_conceded, own_goals

df = geral.query('position == "DF"').copy()
df[['tackles', 'tackles_won', 'blocked_passes', 'interceptions', 'errors']] = defense[['tackles', 'tackles_won', 'blocked_passes', 'interceptions', 'errors']]
df[['aerials_won_pct', 'ball_recoveries', 'pens_conceded', 'own_goals']] =  misc[['aerials_won_pct', 'ball_recoveries', 'pens_conceded', 'own_goals']]
df
```
